```python
#Excercise 1
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('Data1.csv')
df.rename( columns={'Unnamed: 0':'Date'}, inplace=True )
df.set_index('Date')
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(10,10))
plt.plot(df['Date'], df['theta_1'], label = 'theta_1')
plt.plot(df['Date'], df['theta_2'], label = 'theta_2')
plt.plot(df['Date'], df['theta_3'], label = 'theta_3')
plt.plot(df['Date'], df['theta_4'], label = 'theta_4')
plt.plot(df['Date'], df['theta_5'], label = 'theta_5')
plt.plot(df['Date'], df['theta_6'], label = 'theta_6')
plt.legend()
plt.xlabel('date')
plt.ylabel('angle')
plt.show()
fig, axs = plt.subplots(3,2, figsize=(10,10))
fig.suptitle('Histograms')
axs[0][0].hist(df['theta_1'], bins = 50)
axs[0][1].hist(df['theta_2'], bins = 50)
axs[1][0].hist(df['theta_3'], bins = 50)
axs[1][1].hist(df['theta_4'], bins = 50)
axs[2][0].hist(df['theta_5'], bins = 50)
axs[2][1].hist(df['theta_6'], bins = 50)
axs[0][0].title.set_text('theta_1')
axs[0][1].title.set_text('theta_2')
axs[1][0].title.set_text('theta_3')
axs[1][1].title.set_text('theta_4')
axs[2][0].title.set_text('theta_5')
axs[2][1].title.set_text('theta_6')
plt.show()
plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
df['theta_1'].plot.kde()
plt.title("theta_1")
plt.subplot(3,2,2)
df['theta_2'].plot.kde()
plt.title("theta_2")
plt.subplot(3,2,3)
df['theta_3'].plot.kde()
plt.title("theta_3")
plt.subplot(3,2,4)
df['theta_4'].plot.kde()
plt.title("theta_4")
plt.subplot(3,2,5)
df['theta_5'].plot.kde()
plt.title("theta_5")
plt.subplot(3,2,6)
df['theta_6'].plot.kde()
plt.title("theta_6")
plt.show()
df2 = df[(df['Date'] >= '2018-01-01') & (df['Date'] <= '2018-12-31')]
df2['theta_1-theta_4'] = df2['theta_1'] - df2['theta_4']
plt.figure(figsize=(10,10))
plt.plot(df2['Date'], df2['theta_1-theta_4'])
plt.title('theta_1-theta_4')
plt.show()
plt.figure(figsize=(10,10))
plt.hist(df2['theta_1-theta_4'], bins = 50)
plt.title('Histogram')
plt.show()
plt.figure(figsize=(10,10))
df2['theta_1-theta_4'].plot.kde()
plt.title('KDE')
plt.show()




```
